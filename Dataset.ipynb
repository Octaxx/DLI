```python
import pandas as pd

url = "https://raw.githubusercontent.com/Octaxx/DLI/main/CEAS_08.csv"
df = pd.read_csv(url)

counts = df['label'].value_counts()

# Show a few phishing emails (label = 1)
print("Phishing emails examples:\n")
display(df[df['label'] == 1][['subject', 'body']].head(3))
print("\n")

# Show a few safe emails (label = 0)
print("\nSafe emails examples:\n")
display(df[df['label'] == 0][['subject', 'body']].head(3))
print("\n")

# Drop missing values
df = df.dropna()

# Check for remaining missing values
print("Missing values after dropna():\n")
print(df.isnull().sum())
print("\n")  # Another blank line for spacing

# Print total rows
print(f"Total rows after cleaning: {len(df)}\n")

# Show first 100 rows as a table
df.head(100)

```
